3. NumPy Array

The main object in NumPy is the ndarray (N-dimensional array).

Example:

```
import numpy as np

a = np.array([1, 2, 3, 4])

print(a)
```

Output:

[1 2 3 4]

Unlike Python lists, NumPy arrays:

Store elements of the same data type
Support vectorized operations
Are faster for calculations
