What is Matplotlib?

Matplotlib is a popular Python library used for data visualization. It helps us create different types of graphs and charts to represent data in a simple and attractive way.

Developed by: John D. Hunter (2003)

Main Module: matplotlib.pyplot

Features of Matplotlib
Easy to create graphs and charts
Supports 2D plotting
Can create simple 3D plots
Highly customizable
Works well with NumPy and Pandas
Can save graphs in different image formats (PNG, JPG, PDF, SVG)
Used in data science, machine learning, business analytics, and scientific research

Advantages of Matplotlib

Free and open source
Easy to learn
High-quality visualizations
Many chart types available
Integrates with NumPy and Pandas
Supports customization of colors, labels, titles, and styles

Disadvantages of Matplotlib

More code is needed compared to some modern libraries
Complex graphs can take longer to create
Not ideal for highly interactive dashboards

Types of Charts

1. Line Chart

Used to show trends over time.

Example

```python
import matplotlib.pyplot as plt

subjects=["Math","Science","English"]
marks=[80,90,85]

plt.bar(subjects,marks)
plt.show()
```

Applications

Student marks
Product sales
Population comparison

3. Horizontal Bar Char

```python
plt.barh(subjects,marks)
```

4. Pie Chart

Shows percentage distribution

Example

```python
import matplotlib.pyplot as plt

marks=[40,30,20,10]
labels=["Math","Science","English","History"]

plt.pie(marks,labels=labels,autopct="%1.1f%%")
plt.show()
```

Applications

Budget distribution
Market share
Population percentage

5. Histogram

Example

```python
import matplotlib.pyplot as plt

data=[10,20,20,30,30,30,40]

plt.hist(data)
plt.show()
```

Applications

Age distribution
Exam score distribution

6. Scatter Plot

Shows relationship between two variables.

Example

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,25,15,30]

plt.scatter(x,y)
plt.show()
```

Applications

Height vs Weight
Salary vs Experience

7. Area Chart

```python
plt.fill_between(x,y)
```

Applications

Population growth
Sales growth

8. Stem Plot

```python
plt.stem(x,y)
```

9. Step Plot

```python
plt.step(x,y)
```

10. Box Plot

Used to identify data spread and outliers.

```python
plt.boxplot(data)
```

11. Error Bar Plot

```python
plt.errorbar(x,y,yerr=2)
```

Multiple Lines

```python
import matplotlib.pyplot as plt

x=[1,2,3]

y1=[2,4,6]
y2=[1,3,5]

plt.plot(x,y1,label="Line1")
plt.plot(x,y2,label="Line2")

plt.legend()

plt.show()
```
