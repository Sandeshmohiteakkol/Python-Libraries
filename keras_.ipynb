What is Keras?

Keras is an open-source, high-level Deep Learning API used to build and train Artificial Intelligence (AI) and Deep Learning (DL) models. It is integrated with TensorFlow and provides a simple interface for creating neural networks.

Developed by: François Chollet

Released: 2015

Backend: TensorFlow (default)

Website: https://keras.io

Features of Keras

- Open-source and free
- Easy to learn and use
- High-level API for Deep Learning
-Built on top of TensorFlow
- Supports CPU, GPU, and TPU
- Fast model development
- Supports Convolutional Neural Networks (CNN)
- Supports Recurrent Neural Networks (RNN)
- Supports Transfer Learning
- Large community support

Advantages of Keras

- Beginner-friendly
- Less coding required
- Easy model building
- Fast prototyping
- Excellent documentation
- Supports deep learning - applications
- Integrates with TensorFlow

Disadvantages of Keras

- Less flexible than low-level TensorFlow
- Limited control over complex operations
- Depends mainly on TensorFlow backend

# Common Keras Functions

| **Function**         | **Purpose**                                 |
| -------------------- | ------------------------------------------- |
| `keras.Sequential()` | Create a sequential model                   |
| `model.add()`        | Add layers to the model                     |
| `Dense()`            | Create a fully connected (Dense) layer      |
| `model.compile()`    | Configure the model before training         |
| `model.fit()`        | Train the model using training data         |
| `model.evaluate()`   | Evaluate the model performance on test data |
| `model.predict()`    | Predict the output for new input data       |
| `model.save()`       | Save the trained model to a file            |
| `load_model()`       | Load a previously saved model               |
| `keras.Input()`      | Define the input layer of the model         |

# Keras vs TensorFlow

| **Keras**                                            | **TensorFlow**                                             |
| ---------------------------------------------------- | ---------------------------------------------------------- |
| High-level API                                       | Complete Deep Learning Framework                           |
| Easy to learn and use                                | More advanced and feature-rich                             |
| Requires less coding                                 | Requires more coding                                       |
| Fast model development                               | More flexible and customizable                             |
| Best for beginners                                   | Best for complex AI applications                           |
| Runs on TensorFlow                                   | Includes Keras as its official high-level API              |
| Focuses on simplicity                                | Provides complete control over deep learning models        |
| Ideal for rapid prototyping                          | Ideal for large-scale production systems                   |
| Easier to build neural networks                      | Supports advanced model customization                      |
| Suitable for small and medium deep learning projects | Suitable for research and enterprise-level AI applications |
